```python
import pandas as pd
```

```python
df = pd.read_excel('/content/assignment_online_customer (1).xlsx')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[['หมายเลขคำสั่งซื้อ', 'ชื่อผู้ใช้ (ผู้ซื้อ)', 'วันที่ทำการสั่งซื้อ', 'ชื่อสินค้า', 'จำนวน', 'ราคาขายสุทธิ', 'จำนวนเงินทั้งหมด' , 'จังหวัด']]
df.head()
```

```python
df = df.rename(  {'หมายเลขคำสั่งซื้อ'    :  'Transaction_ID'  ,
                           'ชื่อผู้ใช้ (ผู้ซื้อ)'        : 'User_ID',
                           'วันที่ทำการสั่งซื้อ'    : 'Date',
                            'ชื่อสินค้า'              : 'Product_ID',
                            'จำนวน'                 : 'Quantity',
                            'ราคาขายสุทธิ'        : 'Price',
                            'จำนวนเงินทั้งหมด'  : 'Amount',
                            'จังหวัด'                 : 'Province'
                          }  , axis='columns' )
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Date'] = df['Date'].astype('datetime64')
```

```python
df.dtypes
```

```python
import datetime as dt
```

```python
df['buy_day'] = df['Date'].dt.day_name()
```

```python
df.head()
```

```python
df['buy_hour'] = df['Date'].dt.hour
```

```python
df.head()
```

```python
pt_transaction = pd.pivot_table(data=df, index='Transaction_ID' , values='Amount', aggfunc='max' )
pt_transaction = pt_transaction.sort_values(by='Amount' , ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar( data=pt_transaction ,  x=pt_transaction.index , height='Amount')
plt.title('Top 10 highest spending of transaction')
plt.xticks(rotation = 90);
```

```python
pt_product = pd.pivot_table(data=df, index='Product_ID' , values='Price', aggfunc='sum' )
pt_product = pt_product.sort_values(by='Price' , ascending=False).head(10)
```

### Plot graph with Matplotlib

```python
plt.bar( data=pt_product ,  x=pt_product.index , height='Price', color='lightblue')
plt.title('Top 10 highest spending of product')
plt.xticks(rotation = 90);
```

# RMF Analystis

```python
rmf_m = df.groupby('User_ID')['Price'].sum().reset_index()
rmf_m.columns = ['User_ID', 'Monetory']
rmf_m.head(10)
```

```python
rmf_f = df.groupby('User_ID')['Transaction_ID'].count().reset_index()
rmf_f.columns = ['User_ID', 'Frequency']
rmf_f.head(10)
```

```python
rmf = rmf_m.merge( rmf_f, how='inner' , on='User_ID' )
rmf.head(10)
```
